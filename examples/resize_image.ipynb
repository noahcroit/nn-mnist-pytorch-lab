```python
import cv2

image = cv2.imread('img/handwritten_7.jpeg')
resized_image = cv2.resize(image, (28, 28), interpolation=cv2.INTER_AREA)
print("original size=", image.shape, "new size=", resized_image.shape)
print("original=",resized_image[:,:,0])
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(image)
plt.title("original")
plt.show()

plt.figure()
plt.imshow(resized_image)
plt.title("resize")
plt.show()
```

```python
# Revert to the same type with MNIST
inverted_image = cv2.bitwise_not(resized_image)
print(inverted_image[:,:,0])
plt.figure()
plt.imshow(inverted_image)
plt.title("invert")
plt.show()
```

```python
# Normalization 
inverted_image = inverted_image / 256
print("original (normalized)=",inverted_image[:,:,0])
plt.figure()
plt.imshow(inverted_image)
plt.title("invert")
plt.show()
```
